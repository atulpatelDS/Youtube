### Generalized Imputation

### Use fillna() function to use mean, median and mode

```python
df_titanic.head()
```

```python
# Impute missing age with median (Numerical Vari  able)
df_titanic['Age'].fillna(df_titanic['Age'].median(), inplace = True)
```

```python
# Impute Embarked with mode (Categorical Variable)
df_titanic['Embarked'].fillna(df_titanic['Embarked'].mode()[0], inplace = True)
```

### Use replace() function to use mean, median and mode

```python
# Import the dataset
df_titanic = pd.read_csv("https://raw.githubusercontent.com/atulpatelDS/Data_Files/master/Titanic/titanic_train.csv")
```

```python
df_titanic["Age"].replace(np.NaN,df_titanic["Age"].mean(),inplace=True)
# df_titanic["Age"].replace(np.NaN,df_titanic["Age"].median(),inplace=True)
```

```python
df_titanic["Embarked"].replace(np.NaN,df_titanic["Embarked"].mode()[0],inplace=True)
```

```python
# Import Dataset
import pandas as pd
df_Saless = pd.read_excel("https://raw.githubusercontent.com/atulpatelDS/Data_Files/master/Feature_Engineering/Missing_Value/Saless.xlsx")
```

```python
df_Saless['Day_Temp'].fillna(df_Saless.groupby(["Store_Type","City_Type"])['Day_Temp']
                             .transform('mean'),inplace=True)
```

```python
df_Saless["Product_Quality"] = df_Saless.groupby(['Store_Type','City_Type']).Product_Quality.transform(lambda x: x.fillna(x.mode()[0]))
```

```python
# Import the dataset
df_titanic = pd.read_csv("https://raw.githubusercontent.com/atulpatelDS/Data_Files/master/Titanic/titanic_train.csv")
```

```python
# fill the missing Age by mean age of Pclass, Sex and Survived
col = ['Survived','Pclass','Sex']
df_titanic['Age'] = df_titanic['Age'].fillna(df_titanic.groupby(col)['Age'].transform('mean'))
```
